# 放電カメラ画像解析

```python
import glob

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import cv2
```

## 明るい背景画像
放電画像をオーバーラップさせて放電箇所を確認するために、明るい背景画像を用意する。axel.kyoto@gmail.comのgoogleドライブにある、20211101_Discharge.pptxから切り出してきたもの。

```python
# 読み込み
img_base = cv2.imread('base_guide_20240329.png')
img_base = cv2.cvtColor(img_base, cv2.COLOR_BGR2RGB)

# リサイズ
print('original\t:', img_base.shape)  # スクリーンショットのため、480×640からずれている
img_base = cv2.resize(img_base, (640, 480))
print('resized\t\t:', img_base.shape)

# ガンマ補正用のテーブル（画像のmeanに応じて変動）
gamma = 1 / np.sqrt(img_base.mean()) * 13
print('gamma :', gamma)
g_table = np.array([((i / 255.0) ** (1 / gamma)) * 255 for i in np.arange(0, 256)]).astype("uint8")
# ガンマ補正
img_base = cv2.LUT(img_base, g_table)

fig, ax = plt.subplots()
ax.imshow(img_base)

fig.savefig('base_gamma.png')
```

各ビデオの全フレームをarrayとして取り出して、各フレームの画素値の合計値が閾値2.2e5を超えるかで放電か否か判定する（合計値のヒストグラムからだいたいで設定した閾値なので、もっとよい方法はあるはず）。ガンマ、コントラストを調整したうえで背景の明るい画像とアルファブレンドして書き出す。

```python
filelist = sorted(glob.glob('data/20231225_1005_21.avi'))
# コントラスト補正用のテーブル
a = 10
c_table = np.array([255.0 / (1 + np.exp(-a * (i - 128) / 255)) for i in np.arange(0, 256)]).astype("uint8")

for file in filelist:
    name = file.split('/')[-1].split('.')[0]
    
    cap = cv2.VideoCapture(file)
    frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    rgb = 3
    arr = np.zeros((frames, height, width, rgb), dtype='uint8')

    for f in range(frames):
        cap.set(cv2.CAP_PROP_POS_FRAMES, f)
        ret, frame = cap.read()
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        arr[f,:,:,:] = frame
    
    sums = arr.astype('int64').sum(axis=(1,2,3))
    flash_idx = [i for i,x in enumerate(sums) if x > 2.2e5]
    # print(flash_idx)
    
    for idx in flash_idx:
        fig, ax = plt.subplots()
        arr_flash = arr[idx]
        
        # ガンマ補正用のテーブル（画像のmeanに応じて変動）
        #gamma = 1 / np.sqrt(arr_flash.mean()) * 13
        gamma = 1.
        g_table = np.array([((i / 255.0) ** (1 / gamma)) * 255 for i in np.arange(0, 256)]).astype("uint8")
        # ガンマ補正
        img_gamma = cv2.LUT(arr_flash, g_table)
        # コントラスト補正 
        img_gamma = cv2.LUT(img_gamma, c_table)
        
        # 背景画像とのアルファブレンディング
        alpha = 0.6
        blended = cv2.addWeighted(img_base, alpha, img_gamma, 1 - alpha, 0)
          
        ax.imshow(blended)
        ax.set(
            title=f'{name}_frame{idx}'
        )
        fig.savefig(f'20240329/blended/{name}_{idx}.png')
        plt.close(fig)
    
    cap.release()
    
    # fig, ax = plt.subplots()
    # ax.hist(sums)
    # ax.set(title=f'{file}', xlim=(0,1e6))
```

```python
# pickup discharge frame
sums = arr.astype('int64').sum(axis=(1,2,3))
flash_idx = sums > 2.2e5  # need to improve

arr_flash = arr[flash_idx]
```

```python
fig, ax = plt.subplots()

ax.imshow(arr_flash[0])
```

```python
img = arr_flash[0].copy()

# gamma 変換
gammas = np.linspace(1, 5, 9)  # 1, 1.5, 2 ... 4.5, 5
gammas 

for gamma in gammas:
    g_table = np.array([((i / 255.0) ** (1 / gamma)) * 255 for i in np.arange(0, 256)]).astype("uint8")
    # fig, ax = plt.subplots()
    # ax.plot(np.arange(0,256), g_table)
    
    img_gamma = cv2.LUT(img, g_table)
    
    fig, ax = plt.subplots()
    ax.imshow(img_gamma)
```

```python
# alpha, beta
def adjust(img, alpha=1.0, beta=0.0):
    # 積和演算を行う。
    dst = alpha * img + beta
    # [0, 255] でクリップし、uint8 型にする。
    return np.clip(dst, 0, 255).astype(np.uint8)

for alpha in np.arange(1,15,2):
    for beta in [30, 60]:
        out = adjust(img, alpha, beta)
        fig, ax = plt.subplots()
        ax.imshow(out)
```
